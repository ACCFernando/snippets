```python
from pptx import Presentation
from pptx.util import Inches
from pptx.enum.shapes import MSO_SHAPE
from pptx.enum.dml import MSO_THEME_COLOR
```

"Crate PPT Presentation to add slides to"
"Similar concept to tkinter where you add widgets to the main screen"

```python
pr1 = Presentation()
pr1.save("tutorial.pptx")
```

"Refresh to the layouts of each slide" <br>
"0 = Title Slide, 1 = Title and Content, 3 = Section Header, etc"

```python
slide1_register = pr1.slide_layouts[0]
```

Add initial slide to presentation

```python
slide1 = pr1.slides.add_slide(slide1_register)
```

Main top Placeholder

```python
title1 = slide1.shapes.title
```

Placeholder = Item in Layout

```python
subtitle = slide1.placeholders[1]
```

Insert Text

```python
title1.text = "ANALYSITRISING"
subtitle.text = "Subscribe to my Channel"
pr1.save("tutorial.pptx")
```

Paert 2: Add 2 more slides - Bullet points and a picture

```python
slide2_register = pr1.slide_layouts[1]
slide2 = pr1.slides.add_slide(slide2_register)
```

Edit Bullet point Slide

```python
title2 = slide2.shapes.title
title2.text = "Now For some Bullet Points"
```

```python
bullet_point_box = slide2.shapes
```

```python
bullet_points_lvl1 = bullet_point_box.placeholders[1]
bullet_points_lvl1.text = "Subscribe"
```

```python
bullet_points_lvl2 = bullet_points_lvl1.text_frame.add_paragraph()
bullet_points_lvl2.text = "to"
bullet_points_lvl2.level = 1
```

```python
bullet_points_lvl2 = bullet_points_lvl1.text_frame.add_paragraph()
bullet_points_lvl2.text = "my"
bullet_points_lvl2.level = 2
```

```python
bullet_points_lvl2 = bullet_points_lvl1.text_frame.add_paragraph()
bullet_points_lvl2.text = "Channel"
bullet_points_lvl2.level = 3
```

```python
pr1.save("tutorial.pptx")
```

Slide 3 <br>
Create Bullet Point and Picture Slide

```python
slide3_register = pr1.slide_layouts[5]#5 = only title layout
slide3 = pr1.slides.add_slide(slide3_register)
```

```python
title3 = slide3.shapes.title
title3.text = "Picture Time"
```

Add Image

Part 3:Auto-shapes <br>
Slide 4 <br>
Register / Create Slide

```python
slide4_register = pr1.slide_layouts[5]
slide4 = pr1.slides.add_slide(slide4_register)
```

Slide 4 Title

```python
title4 = slide4.shapes.title
title4.text = "Shapework"
```

Create Shapes <br>
Shape 1

```python
left1 = top1 = width1 = height1 =  Inches(2)
```

```python
add_shape2 = slide4.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE, 
                                     left1, 
                                     top1, 
                                     width1, 
                                     height1)                          
```

```python
pr1.save("tutorial.pptx")
```

Shape 2

```python
left2 = Inches(6)
top2 = Inches(2)
width2 = height2 = Inches(2)
```

```python
arrow1 = slide4.shapes.add_shape(MSO_SHAPE.DOWN_ARROW,
                                 left2,
                                 top2,
                                 width2,
                                 height2)
```

Change arrow main colour

```python
fill_arrow1 = arrow1.fill
fill_arrow1.solid()
fill_arrow1.fore_color.theme_color = MSO_THEME_COLOR.ACCENT_2
```

```python
pr1.save("tutorial.pptx")
```

Rotate Shape

```python
arrow1.rotation = 90  #-90 gira anti-horario
```

```python
pr1.save("tutorial.pptx")
```
